# K Nearest neighbors Classifier

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
```

```python
data = pd.read_csv('D:/Extra Classes/Machine Learning/DataSets/teleCust1000t.csv')
data.head()
```

```python
data['custcat'].value_counts()
```

```python
data.hist(column = 'income', bins = 50)
```

```python
# pandas dataframe to numpy array
x = data[['region', 'tenure','age', 'marital', 'address', 'income', 'ed', 'employ','retire', 'gender', 'reside']].values
y = data['custcat'].values
print(x[0:5])
print(y[0:5])
```

```python
# lets normalize data
scaler = StandardScaler()
x_normal = scaler.fit_transform(x)
print('x data is:', x)  
print('x normal data is:', x_normal)  
```

```python
# lets split data
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split( x, y, test_size=0.2, random_state=4)
```

```python
# Now lets use KNN Classifier
from sklearn.neighbors import KNeighborsClassifier
# use random k
k = 4
neigh = KNeighborsClassifier(n_neighbors = k)
neigh.fit(x_train, y_train)
```

```python
# predicting
y_hat = neigh.predict(x_test)

# evaluation
from sklearn import metrics
print("Train set Accuracy: ", metrics.accuracy_score(y_train, neigh.predict(x_train)))
print("Test set Accuracy: ", metrics.accuracy_score(y_test, y_hat))
```
